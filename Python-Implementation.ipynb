1)  Extract Text from the PDF 

```python
import pdfplumber

pdf_path = "Eveready Financial Report.pdf"

with pdfplumber.open(pdf_path) as pdf:
    for i, page in enumerate(pdf.pages):
        print(f"\n--- Page {i+1} ---\n")
        print(page.extract_text())  # Extracts text from each page
        print("\n--- End of Page ---\n")
```

2) Extract Tables from the PDF

```python
import pandas as pd

with pdfplumber.open(pdf_path) as pdf:
    for i, page in enumerate(pdf.pages):
        tables = page.extract_tables()
        for table in tables:
            df = pd.DataFrame(table)
            print(f"\n--- Table on Page {i+1} ---\n")
            print(df)  # Displays the table as a Pandas DataFrame
            print("\n")
```

Table extraction is working, but it's messy. Let's clean it up by:
1️. Flattening the table
2️. Handling merged/missing values
3️. Extracting key metrics like Profit Before Tax

```python
import pdfplumber
import pandas as pd

pdf_path = "Eveready Financial Report.pdf"

data_frames = []

with pdfplumber.open(pdf_path) as pdf:
    for i, page in enumerate(pdf.pages):
        tables = page.extract_tables()
        for table in tables:
            df = pd.DataFrame(table)  # Convert to DataFrame
            df = df.dropna(how="all")  # Remove empty rows
            df = df.applymap(lambda x: x.replace("\n", " ") if isinstance(x, str) else x)  # Remove newlines
            data_frames.append(df)
            print(f"\n--- Cleaned Table from Page {i+1} ---\n")
            print(df)

# Save the cleaned tables
with open("cleaned_financial_tables.json", "w") as f:
    f.write(pd.concat(data_frames).to_json(indent=4, orient="records"))
    print("\n--- Cleaned Tables Saved ---\n")
```
